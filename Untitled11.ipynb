```python
from farasa.stemmer import FarasaStemmer
```

```python
stemmer = FarasaStemmer()
```

```python
job1 = input("اسم الوظيفة الاولى :  ")
print(job1)
```

```python
job2 = input("اسم الوظيفة الثانية :  ")
print(job2)
```

```python
stemmed1 = stemmer.stem(job1)
print (stemmed1)
stemmed2 = stemmer.stem(job2)
print (stemmed2)
```

```python
translate1=GoogleTranslator(source='ar',target='en').translate(stemmed1)
print(translate1)
translate2=GoogleTranslator(source='ar',target='en').translate(stemmed2)
print(translate2)
```
